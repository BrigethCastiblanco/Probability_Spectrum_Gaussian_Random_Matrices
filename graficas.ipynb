```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dim = 250
quant = 1000
```

```python
#Carga de datos
 
goe = np.load(f"data/espectro_GOE_dim{dim}_quant{quant}.npy")
gue = np.load(f"data/espectro_GUE_dim{dim}_quant{quant}.npy")
gse = np.load(f"data/espectro_GSE_dim{dim}_quant{quant}.npy")
```
