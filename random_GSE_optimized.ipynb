```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import gc
dim = 25                 #matrix dimension
matrix_quant = 5000      #ensamble dimension
```

```python

all_eigenvalues = []

# Pre-asignamos una sola matriz 
S = np.empty((2*dim, 2*dim), dtype=complex)     #Asignación de la estructura matricial cuaternionica

for m in tqdm(range(matrix_quant), desc="GSE Seguro"):
    # Generamos los componentes (complejos)
    A = np.random.normal(0, 1, (dim, dim)) + 1j * np.random.normal(0, 1, (dim, dim))
    B = np.random.normal(0, 1, (dim, dim)) + 1j * np.random.normal(0, 1, (dim, dim))
    
    # Aplicamos simetrías y normalización (Radio = 2)
    q_a = (A + A.conj().T) / np.sqrt(8 * dim)
    q_b = (B - B.T) / np.sqrt(8 * dim)
    
    # Rellenamos la matriz S pre-asignada siguiendo la estructura matricial cuaternionica
    S[:dim, :dim] = q_a
    S[:dim, dim:] = q_b
    S[dim:, :dim] = -np.conj(q_b)
    S[dim:, dim:] = np.conj(q_a)
    
    # Calculamos valores propios (solo los valores reales)
    evs = np.linalg.eigvalsh(S)
    
    # Guardamos los resultados
    all_eigenvalues.extend(evs)
    
    # Cada 500 matrices, liberamos memoria explícitamente por seguridad
    if m % 500 == 0:
        gc.collect()

# Convertimos a array final y guardamos en disco inmediatamente
values_gse = np.array(all_eigenvalues)
nombre_archivo = f"espectro_GSE_dim{dim}_quant{matrix_quant}.npy"
np.save(f"data/{nombre_archivo}", values_gse)
```
