```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import gc
dim = 25                 #matrix dimension
matrix_quant = 5000      #ensamble dimension
```

```python
all_eigenvalues = []

# Pre-asignamos una sola matriz

for m in tqdm(range(matrix_quant), desc="GOE Seguro"):
    # Generamos los componentes
    r = np.random.normal(0, 1, (dim, dim))
    
    # Aplicamos simetrías y normalización (Radio = 2)
    O_T = r.T                  #tomar dagga de los elementos
    symmetric_O = (r + O_T)/np.sqrt(2*dim)      #normalización

    evs = np.linalg.eigvalsh(symmetric_O)       #Valores propios
    
    # Guardamos solo los resultados numéricos
    all_eigenvalues.extend(evs)
    
    # Cada 500 matrices, liberamos memoria explícitamente por seguridad
    if m % 500 == 0:
        gc.collect()

# Convertimos a array final y guardamos en disco inmediatamente
values_goe = np.array(all_eigenvalues)
nombre_archivo = f"espectro_GOE_dim{dim}_quant{matrix_quant}.npy"
np.save(f"data/{nombre_archivo}", values_goe)
```
