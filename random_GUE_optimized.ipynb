```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import gc
dim = 25                 #matrix dimension
matrix_quant = 5000      #ensamble dimension
```

```python
all_eigenvalues = []

# Pre-asignamos una sola matriz

for m in tqdm(range(matrix_quant), desc="GUE Seguro"):
    # Generamos los componentes complejas
    z = np.random.normal(0, 1, (dim, dim)) + 1j * np.random.normal(0, 1, (dim, dim))
    
    # Aplicamos simetrías y normalización (Radio = 2)
    z_dag = z.conj().T                  #Tomando el dagga de los elementos
    Unitary_hermite = (z + z_dag)/np.sqrt(4*dim)    #matriz compleja hermítica

    evs = np.linalg.eigvalsh(Unitary_hermite)   #Valores propios
    
    # Guardamos solo los resultados
    all_eigenvalues.extend(evs)
    
    # Cada 500 matrices, liberamos memoria explícitamente por seguridad
    if m % 500 == 0:
        gc.collect()

# Convertimos a array final y guardamos en disco inmediatamente
values_gue = np.array(all_eigenvalues)
nombre_archivo = f"espectro_GUE_dim{dim}_quant{matrix_quant}.npy"
np.save(f"data/{nombre_archivo}", values_gue)
```
